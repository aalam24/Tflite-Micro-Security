```python
import tensorflow.keras
from tensorflow.keras import models, layers, optimizers
import numpy as np
x = np.linspace(1,4,400)
y = (x*4) + (np.random.randn(x.shape[0]) * .75)
print("x:", x)
print("y:", y)
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, ".")
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Dense(1, input_dim = 1, activation="linear"))
print(model.layers[0].get_weights())
```

```python
model.compile(optimizer = optimizers.SGD(), loss = "mse", metrics = ["mse"])
model.fit(x, y, batch_size = 1, epochs = 400, shuffle = False)
```

```python
model.predict([6.9])
```

```python
y_predictions = model.predict(x)
plt.plot(x, y, ".", x, y_predictions, "r")
plt.show()
```
